```python
import matplotlib.pyplot as plt
import math

# La ecuacin de la reaccin global es: Zn(s) + Cu2^+(aq) ---> Zn^2+(aq) + Cu(s)

E0 = 1.10
F = 96485
n = 2
R = 8.314
T = 298
c0 = 1
m = 0.1 # Aclaracion m = masa de electrodos
M_Zn = 65.38
M_Cu = 63.55

def voltaje(c):
    E = E0 - (R*T)/(n*F) * math.log(c/(c0-c))
    return E

def corriente(E):
    R = 10
    I = E/R
    return I

def concentracion(t):
    I = 0.05
    c = (I*t)/(n*F)
    return c

t_max = 10*3600
t_list = range(1, t_max, 60)

E_list = []
I_list = []
c_list = []

for t in t_list:
    c = concentracion(t)
    E = voltaje(c)
    I = corriente(E)

    c_list.append(c)
    E_list.append(E)
    I_list.append(I)

plt.plot(t_list, E_list, label="Voltaje")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.title("Variacin del voltaje de la celda galvnica")
plt.legend()
plt.show()

plt.plot(t_list, I_list, label="Corriente")
plt.xlabel("Tiempo (s)")
plt.ylabel("Corriente (A)")
plt.title("Variacin de la corriente de la celda galvnica")
plt.legend()
plt.show()

plt.plot(t_list, c_list, label="Concentracin de Zn^2+")
plt.xlabel("Tiempo (s)")
plt.ylabel("Concentracin (mol/L)")
plt.title("Variacin de la concentracin de Zn2+ en la celda galvnica")
plt.legend()
plt.show()
```
