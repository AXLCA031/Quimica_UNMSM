```python
#El voltaje aumenta a menor concentración de electrolitos porque el potencial del electrodo disminuye cuando aumenta la concentracion de los iones que se oxidan o reducen.
#Esto se debe a que al haber mas iones disponibles, la reacción se desplaza a la derecha y el electrodo pierde tendencia a ganar o perder electrones.
import numpy as np
import matplotlib.pyplot as plt

electrolitos = ["Electrolito A", "Electrolito B", "Electrolito C", "Electrolito D"]
concentraciones = [0.2, 0.1, 0.05, 0.03]

def voltaje_electrolisis(electrolito, concentracion):
  E0 = 1.23
  R = 8.314
  T = 298
  n = 2
  F = 96485
  Q_dict = {"Electrolito A": concentracion,
            "Electrolito B": concentracion,
            "Electrolito C": concentracion,
            "Electrolito D": concentracion}
  Q = Q_dict[electrolito]
  E = E0 - R * T  * np.log(Q) / (n * F)

  return E

def eficiencia_electrolisis(electrolito, concentracion):
  n1 = concentracion
  R = 0.082
  T = 298
  P = 1
  V_teorico = n1 * R * T / P
  z = 2
  F = 96485
  I = 0.5
  t = 3600
  n2 = (I * t)/(F * z)
  V_producido = n2 * R * T / P
  eficiencia = 100 * (V_producido / V_teorico)
  return eficiencia

def oxigeno_electrolisis(electrolito, concentracion):
  n = concentracion / 2
  R = 0.0821
  T = 298
  P = 1
  V = n * R * T / P
  return V

voltajes = []

eficiencias = []

oxigenos = []

for electrolito, concentracion in zip(electrolitos, concentraciones):
  voltaje = voltaje_electrolisis(electrolito, concentracion)
  voltajes.append(voltaje)
  eficiencia = eficiencia_electrolisis(electrolito, concentracion)
  eficiencias.append(eficiencia)
  oxigeno = oxigeno_electrolisis(electrolito, concentracion)
  oxigenos.append(oxigeno)

plt.plot(electrolitos, voltajes, color="red", marker = "o", label="Voltaje (V)")
plt.xlabel("Electrolito")
plt.ylabel("Voltaje (V)")
plt.title("Voltaje de la electrolisis del agua con diferentes electrolitos")
plt.legend()
plt.show()

plt.plot(electrolitos, eficiencias, color="red", marker = "o", label="Eficiencia (%)")
plt.xlabel("Electrolito")
plt.ylabel("Eficiencia (%)")
plt.title("Eficiencia de la electrolisis del agua con diferentes electrolitos")
plt.legend()
plt.show()

plt.bar(electrolitos, oxigenos, label="Oxígeno (L)")
plt.ylim(0,10)
plt.xlabel("Electrolito")
plt.ylabel("Volumen (L)")
plt.title("Volumen del oxígeno producidos en la electrolisis del agua con diferentes electrolitos")
plt.legend()
plt.show()
```
